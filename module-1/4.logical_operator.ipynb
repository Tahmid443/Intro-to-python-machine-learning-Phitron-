## Logical Operator
# 'and' (if one false whole statement will false)
# 'or' (if one true whole statement will true)

```python
x = 10
y = 5
z = 3
result = (x>y) and (x>z)
print(result)
```

```python
x = 10
y = 5
z = 3
result = (x > y) or (x > z)
print(result)
```
