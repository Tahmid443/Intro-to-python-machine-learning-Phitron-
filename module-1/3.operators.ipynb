## Arithmetic Operator

```python
x = 10
y = 3
sum = x + y
sub = x - y
mult = x * y
div = x / y
rem = x % y
print(sum, sub, mult, div, rem)
```

## Comparison Operator

```python
x = 10
y = 5
greater_than = x > y
greater_than_equal = x >= y
less_than = x < y
less_than_equal = x <= y
isEqual = x==y
notEqual = x!=y
print(greater_than,greater_than_equal,less_than,less_than_equal,isEqual,notEqual)
```
