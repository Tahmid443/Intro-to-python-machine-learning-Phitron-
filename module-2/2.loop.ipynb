## Loop in python

## For Loop

```python
for i in range(10):
    i = i+1
    print(i,"Hello Tahmid")
```

```python
#range(start, end, step)
for i in range(1,10,2):
    print(i)
```

```python
for i in range(1,6):
    for j in range(1,5):
        print(i,j)
```

## While Loop

```python
count = 1
while count<10:
    print(count, "Hello Tahmid")
    count = count+1
print(count)
```

```python
count = 1
while True:
    print("This is an infinity loop")
```

```python
count = 1
while False:
    print("This will not execute")
```
