## Break and Continue

```python
sum = 0
for i in range(1,51):
    if i%2==0:
        continue
    sum+=i   
print(sum)
```

```python
accuracy = 95
for i in range(20):
    accuracy+=1
    print(accuracy)
    if accuracy == 100:
        break
    
```

```python
accuracy = 95
while True:
    if accuracy == 100:
        break
    accuracy+=1
    print(accuracy)
```
